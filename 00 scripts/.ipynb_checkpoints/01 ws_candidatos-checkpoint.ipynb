## PARTIDOS -- PRESIDENCIALES

```python
def get_process_data(list_dfs):
    if isinstance(list_dfs,list):
        df = pd.concat(list_dfs).reset_index(drop=True)    
        df = df.dropna(axis=1, how='all')
    else:
        df = list_dfs.dropna(axis=1, how='all')        
    nunique = df.apply(pd.Series.nunique)
    cols_to_drop = nunique[nunique == 1].index
    df = df.drop(cols_to_drop, axis=1)
    return df

```

```python
r_partidos_presi = requests.get('https://plataformaelectoral.jne.gob.pe/Candidato/GetExpedientesLista/110-1-------0-')
json_partidos_presi = json.loads(r_partidos_presi.text)
```

```python
data_partidos_presi = json_partidos_presi['data']
df_partidos_presi = pd.DataFrame(data_partidos_presi)
df_partidos_presi = df_partidos_presi.drop('listaCandidato',axis=1)
```

```python
df_partidos_presi = get_process_data(df_partidos_presi)
```

```python
df_partidos_presi.to_csv(ruta_input + '/base_partidos_presidencial.csv',index=False,encoding='latin-1')
```

```python
def get_dict_data(list_data_hdv_candidatos,list_keys_data):

    dict_list_dfs = {}

    for key_data in list_keys_data:
        dict_list_dfs[key_data] = []

    for data_hdv_candidato in list_data_hdv_candidatos:
        for key_data in list_keys_data:
            if isinstance(data_hdv_candidato[key_data],list):
                df = pd.DataFrame(data_hdv_candidato[key_data]) 
            else:
                df = pd.DataFrame([data_hdv_candidato[key_data]]) 
            dict_list_dfs[key_data].append(df)
    
    return dict_list_dfs
    
```

```python
def save_data(dict_data_hdv,list_keys_data,ruta_input,str_id=''):
    for key_data in list_keys_data:
        print(key_data)
        df = get_process_data(dict_data_hdv[key_data])
        df.to_csv(ruta_input+'/data/base_candidatos{}_{}.csv'.format(str_id,key_data[1:]),index=False,encoding='utf-8-sig')
```

```python
list_keys_data = ['oDatosPersonales','lExperienciaLaboral','oIngresos',
                  'lBienInmueble','lBienMueble','oInfoAdicional',
                  'lSentenciaPenal', 'lSentenciaObliga',
                  'oEduBasica', 'oEduTecnico', 'oEduNoUniversitaria', 'lEduUniversitaria', 'oEduPosgrago',
                  'lCargoPartidario', 'lCargoEleccion', 'lRenunciaOP']

dict_data_hdv_presi = get_dict_data(list_data_hdv_candidatos_presi,list_keys_data)
```

```python
save_data(dict_data_hdv_presi,list_keys_data,ruta_input,'_presi')
```

## PLAN DE GOBIERNO -- RESUMEN

```python
list_partido_politico = df_partidos_presi['strOrganizacionPolitica'].tolist()
list_id_plan_gobierno = df_partidos_presi['idPlanGobierno'].tolist()
```

```python
list_df_propuestas = []
list_df_prop_metadata = []
for partido_politico, id_plan_gobierno in zip(list_partido_politico,list_id_plan_gobierno):
    print(partido_politico + ' ' + str(id_plan_gobierno))
    r_plan_gobierno = requests.get('https://plataformaelectoral.jne.gob.pe/Candidato/GetPlanGobiernoById/{}'.format(
                    id_plan_gobierno))
    json_plan_gobierno = json.loads(r_plan_gobierno.text)
    data_plan_gobierno = json_plan_gobierno['data']
    
    meta_data_columns = list(data_plan_gobierno.keys())[:-5]
    propuestas_columns = list(data_plan_gobierno.keys())[-5:-1]

    meta_data_plan_gobierno = { meta_key: data_plan_gobierno[meta_key] for meta_key in meta_data_columns }
    prop_data_plan_gobierno = { prop_key: data_plan_gobierno[prop_key] for prop_key in propuestas_columns }

    df_prop_metadata = pd.DataFrame([meta_data_plan_gobierno])
    list_df_prop_metadata.append(df_prop_metadata)
    
    list_df_props = []
    for key in prop_data_plan_gobierno.keys():
        df_prop = pd.DataFrame(prop_data_plan_gobierno[key])
        df_prop['strTipoPropuesta'] = key
        list_df_props.append(df_prop)
    df_propuestas = pd.concat(list_df_props).reset_index(drop=True)
    df_propuestas['idPlanGobierno'] = id_plan_gobierno
    list_df_propuestas.append(df_propuestas)
```

```python
df_prop_metadata_partido = get_process_data(list_df_prop_metadata)
df_propuestas_partido = get_process_data(list_df_propuestas)
```

```python
df_propuestas_partido.head()
```

```python
df_prop_metadata_partido.to_excel(ruta_input + '/plan_gobierno_res/propuestas_metadata.xlsx',index=False,encoding='latin-1')
df_propuestas_partido.to_excel(ruta_input + '/plan_gobierno_res/propuestas_resumen_partido.xlsx',index=False,encoding='latin-1')
```
